I think we will need to pull many different topics from the last week - a bit of a worry is that we will just have loads of tweets without much attention and the user itself will be the only really important factor to take into account

```python
def get_topic_tweets(keyword, search_type='search', start_date='2021-01-01', limit=None, return_df=True):
    c = twint.Config()
    
    c.Hide_output = True
    
    if limit:
        c.Limit = limit
        
    c.Since = start_date
    
    if search_type == 'search':
        c.Search = keyword
    else:
        c.Username = "sanchezcastejon"
    
    if return_df:
        c.Pandas = True
        twint.run.Search(c)
        return twint.storage.panda.Tweets_df
```

```python
tweets = get_topic_tweets('machine learning', limit=1000)
```

```python
tweets.groupby('video')[['nlikes', 'nreplies', 'nretweets']].describe()
```
